```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
hpsa_df = pd.read_csv("resources/HPSA.csv")
hpsa_df.head()
```

```python
obesity_df = pd.read_csv("resources/Cities_Obesity.csv")
obesity_df.head()
```

```python
obesity_df = obesity_df.groupby(['CityName', 'StateAbbr']).mean()['Data_Value'].reset_index()
obesity_df
```

```python
hpsa_df.dropna(subset=['HPSA City'], inplace=True)
hpsa_df[['HPSA Score', '% of Population Below 100% Poverty', 'HPSA City']].tail(60)
```

```python
hpsa_df = hpsa_df[['HPSA City', 'HPSA Score']]
hpsa_df = hpsa_df.groupby('HPSA City').mean().reset_index()
hpsa_df
```

```python
merge_df = pd.merge(obesity_df, hpsa_df,left_on='CityName', right_on='HPSA City')
merge_df
```

```python
xs = merge_df['HPSA Score']
ys = merge_df['Data_Value']

plt.scatter(xs, ys)
plt.xlabel('HPSA Score')
plt.ylabel('Obesity %')
plt.show()
```
